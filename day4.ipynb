```python
print('\033[33m' + " ===YOUR ADVENTURE BEGINS HERE=== " + '\033[0m')

# ANSI Color Codes
RED = "\033[31m"
YELLOW = "\033[33m"
GREEN = "\033[32m"
CYAN = "\033[36m"
WHITE = "\033[37m"
RESET = "\033[0m"

print(WHITE + "Welcome to Justice League: Choose Your Adventure!" + RESET)
print(CYAN + "Choose your hero:" + RESET)
print(RED + "1. Batman     (Strength 2, Intellect 6, Speed 2)" + RESET)
print(GREEN + "2. Wonder Woman (Strength 5, Intellect 2, Speed 3)" + RESET)
print(YELLOW + "3. The Flash   (Strength 1, Intellect 2, Speed 7)" + RESET)
print(CYAN + "4. Green Lantern (Strength 3, Intellect 4, Speed 3)" + RESET)
print(WHITE + "5. Superman    (Strength 5, Intellect 3, Speed 2)" + RESET)

choice = input("Enter the number of your hero: ")

if choice == "1":
  hero = "Batman"
  strength = 2
  intellect = 6
  speed = 2
elif choice == "2":
  hero = "Wonder Woman"
  strength = 5
  intellect = 2
  speed = 3
elif choice == "3":
  hero = "The Flash"
  strength = 1
  intellect = 2
  speed = 7
elif choice == "4":
  hero = "Green Lantern"
  strength = 3
  intellect = 4
  speed = 3
elif choice == "5":
  hero = "Superman"
  strength = 5
  intellect = 3
  speed = 2
else:
  print(RED + "Invalid choice. Game over." + RESET)
  exit()

print(GREEN + "You chose " + hero + "!" + RESET)

# Villain selection with full name mapping
print(YELLOW + "Choose your villain:" + RESET)
print(RED + "A. Zod\nB. Lex Luthor\nC. Joker\nD. Sinestro" + RESET)
villain_choice = input("Enter the letter of your villain: ").upper()

if villain_choice == "A":
  villain = "Zod"
elif villain_choice == "B":
  villain = "Lex Luthor"
elif villain_choice == "C":
  villain = "Joker"
elif villain_choice == "D":
  villain = "Sinestro"
else:
  print(RED + "Invalid villain. Game over." + RESET)
  exit()

print(CYAN + "The world is under attack by " + villain + "!" + RESET)
print("You, " + hero + ", spring into action!")

print(WHITE + "Do you want to sneak in or go head-on? (sneak/head-on)" + RESET)
approach = input().lower()

if approach == "sneak":
  if intellect >= 5:
    print(GREEN + "You outsmart the traps and gather intel." + RESET)
  else:
    print(RED + "You trigger a trap! You're injured!" + RESET)
elif approach == "head-on":
  if strength >= 5:
    print(GREEN + "You overpower the enemy forces!" + RESET)
  else:
    print(RED + "You're overwhelmed by enemy fire!" + RESET)
else:
  print(RED + "You hesitated. The villain escapes!" + RESET)
  exit()

print(WHITE + "Final challenge! Use your speed to defuse a bomb. (yes/no)" +
      RESET)
final = input().lower()

if final == "yes":
  if speed >= 5:
    print(GREEN + "You defuse the bomb just in time. Earth is saved!" + RESET)
  else:
    print(RED + "Too slow! The bomb explodes. Game over." + RESET)
else:
  print(RED + "You chose not to act. The world suffers..." + RESET)

print(YELLOW + "Thanks for playing, " + hero + "!" + RESET)
```
